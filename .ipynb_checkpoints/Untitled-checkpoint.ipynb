## Dataset Explanation
This data set contains 113,937 loans with 81 variables on each loan, including loan amount, borrower rate (or interest rate), current loan status, borrower income, and many others.
This data dictionary explains the variables in the data set.
You are not expected to explore all of the variables in the dataset! Focus your exploration on about 10-15 of them.

## Questions
- What factors affect a loan’s outcome status?
- What affects the borrower’s APR or interest rate?
- Are there differences between loans depending on how large the original loan amount was?

```python
df_working = df_prosper[['LoanStatus','BorrowerRate','Term','CreditGrade','EmploymentStatus','EmploymentStatusDuration',
                    'IsBorrowerHomeowner','DebtToIncomeRatio','IncomeRange','IncomeVerifiable','LoanOriginalAmount',
                    'MonthlyLoanPayment','TotalCreditLinespast7years']]
df_working= df_working.copy()
```

null values (nan) should be converted to 'Not available'

```python
df_working.EmploymentStatus.fillna(value='Not available', inplace=True)
df_working.EmploymentStatus.unique()
```

```python
df_working.IsBorrowerHomeowner.unique()
```

```python
df_working.IncomeRange.unique()
```

'Not employed' and '$0' have the same meaning.

```python
df_working.IncomeRange.replace('Not employed', '$0', inplace=True)
```

```python
df_working.IncomeRange.unique()
```

```python
df_working.IncomeVerifiable.unique()
```

```python
df_working.TotalCreditLinespast7years= df_working.TotalCreditLinespast7years.astype('Int64')
df_working.info()
```

```python
df_working.LoanStatus.unique()
```

```python
sns.countplot(df_working.LoanStatus, color='0.5', order=df_working.LoanStatus.value_counts().index);
plt.xticks(rotation=90);
```

```python
# Cut the window in 2 parts
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
 
# Add a graph in each part
sns.boxplot(df_working["BorrowerRate"], ax=ax_box)
sns.distplot(df_working["BorrowerRate"], ax=ax_hist)
 
# Remove x axis name for the boxplot
ax_box.set(xlabel='');
```

```python
sns.jointplot('EmploymentStatusDuration','MonthlyLoanPayment',data=df_working,color='k');
```

```python
import statsmodels.api as sm
log_m = sm.Logit(df_working['LoanStatus'], df_working[['BorrowerRate','Term','CreditGrade','EmploymentStatus','EmploymentStatusDuration',
                    'IsBorrowerHomeowner','DebtToIncomeRatio','IncomeRange','IncomeVerifiable','LoanOriginalAmount',
                    'MonthlyLoanPayment','TotalCreditLinespast7years']])
log_results= log_m.fit()
```
