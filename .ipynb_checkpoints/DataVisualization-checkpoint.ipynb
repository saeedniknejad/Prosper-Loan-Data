## Univariate Exploration

Let's start our exploration by looking at the main variables which are the subject of the questions: LoanStatus, BorrowerAPR and LoanOriginalAmount. Are their distribution skewed or symmetric? Are they unimodal or multimodal? How about the categorical data; LoanStatus?

```python
# univariate plot of LoanStatus as a Categorical data 
sns.countplot(df_loan['LoanStatus'], order = df_loan.LoanStatus.value_counts().index ,color = '0.5' )
plt.xticks(rotation = 90);
```

**LoanStatus** Comments:
- Almost all the data can be placed in the first four categories (Current, Completed, Chargedoff, Defaulted).
- Current category contains more loans than any others.

```python
# univariate plot of BorrowerAPR
# Median and Mean comparison 
median_APR= df_loan.BorrowerAPR.median()
mean_APR= df_loan.BorrowerAPR.mean()



# Cut the window in 2 parts
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
 
# Add a graph in each part
sns.boxplot(df_loan["BorrowerAPR"], ax=ax_box)
#sns.distplot(df_loan["BorrowerAPR"], ax=ax_hist)

sns.distplot(df_loan['BorrowerAPR'],bins=30,kde=True).annotate('''Median = {}
Mean = {}'''.format(median_APR, mean_APR), (0.3,4.3))
plt.axvline(x=median_APR, color='red')
plt.axvline(x=mean_APR, color='black');

# Remove x axis name for the boxplot
ax_box.set(xlabel='');
```

**BorrowerRate** Distribution Comments:
- All values >0.45 can be considered as outliers, based on the Box plot
- Mean and Median are very close to each other, Mean is slightly larger than Median
- It is a slightly right skewed distribution with a spike around 0.35

```python
# univariate plot of LoanOriginalAmount 
# Median and Mean comparison 
median_LOA= df_loan.LoanOriginalAmount.median()
mean_LOA= df_loan.LoanOriginalAmount.mean()

# Cut the window in 2 parts
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
 
# Add a graph in each part
sns.boxplot(df_loan["LoanOriginalAmount"], ax=ax_box)

sns.distplot(df_loan['LoanOriginalAmount'],bins=30,kde=True).annotate('''Median = {}
Mean = {}'''.format(median_LOA, mean_LOA), (20000,0.00012))
plt.axvline(x=median_LOA, color='red')
plt.axvline(x=mean_LOA, color='black');

# Remove x axis name for the boxplot
ax_box.set(xlabel='');
```

```python
# univariate plot of LoanOriginalAmount 
# Cut the window in 2 parts
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
 
# Add a graph in each part
mybins=np.logspace(0, np.log(100), 100)
sns.boxplot(df_loan["LoanOriginalAmount"], ax=ax_box)
sns.distplot(df_loan["LoanOriginalAmount"], ax=ax_hist, bins=mybins)
mybins=np.logspace(0, np.log(100), 100)
plt.xscale('log')

 
# Remove x axis name for the boxplot
ax_box.set(xlabel='');
```

**LoanOriginalAmount** Distribution Comments:
- All values >25000 can be considered as outliers, based on the Box chart
- Mean and Median are quite different, Mean is larger than Median
- Data shows positively skewed distribution with majority of datapoints falling between 1,000 and 10,000 
- Based on the logarithmic view, data is quite spiky and It's very hard to call it as Unimodal or even Bimodal

## Bivariate Exploration

Above, I looked at the univariate distribution of three features in the **Prosper Loan** dataset: LoanStatus, BorrowerAPR and LoanOriginalAmount. Here, I'll investigate relationships between pairs of these variables, particularly how each of them relate to the three above mentioned features.


```python
# BorrowerAPR variation for each LoanStatus category
sns.boxplot(data=df_loan, x='LoanStatus', y='BorrowerAPR', color='0.5', order=df_loan.LoanStatus.value_counts().index)
plt.xticks(rotation=90);
```

**BorrowerAPR for LoanStatus categories** insights:

- Considering the bar chart from univariate exploration of LoanStatus, we know that major LoanStatus categories are established to be Current, Completed, Chargedoff and Defaulted. Therefore, we have used `order` keyword in the box plot.
- Mean values of BrrowerAPR for Current, Completed LoanStatus categories are significantly lower than Chargedoff and Defaulted categories. 

```python
# BorrowerAPR variation versus BorrowRate
sns.jointplot('BorrowerRate','BorrowerAPR',data=df_loan).annotate(stats.pearsonr);
```

**BorrowerAPR vs. BorrowerRate** insight:
- BorrowAPR is related to Borrower rate. 
-There is a strong positive relationship between the 2 variables. Pearson r of 0.99 is indicative of this fact.

## Multivariate Exploration

So far, we looked at various univariate and bivariate relationships. We saw that the log of price was approximately linearly related to the cube root of carat weight, as analogy to its length, width, and depth. You also saw that there was an unintuitive relationship between price and the categorical quality measures of cut, color, and clarity, that the median price decreased with increasing quality. Investigating the distributions more clearly and looking at the relationship between carat weight with the three categorical variables showed that this was due to carat size tending to be smaller for the diamonds with higher categorical grades.

The goal of this workspace will be to depict these interaction effects through the use of multivariate plots.

To start off with, create a plot of the relationship between price, carat, and clarity. In the previous workspace, you saw that clarity had the clearest interactions with price and carat. How clearly does this show up in a multivariate visualization?

```python
sns.scatterplot(x='BorrowerAPR', y='AvailableBankcardCredit', data=df_loan, hue='CreditGrade');
plt.yscale('log')
plt.ylim(1, 1000000);
```
