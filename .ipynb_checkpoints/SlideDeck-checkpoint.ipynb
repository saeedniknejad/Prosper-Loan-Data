### Basic Cleaning:

Our main goal for cleaning is to enable the plotting libraries plot the data without problems. It is also imparative to remove bad/missing data to avoid misleading plots.

**Define:** 

- change EmploymentStatus from nan to 'Not Available'
- change IncomeRange from Not Employed to $0 
- TotalCreditLinespast7years type is float, change to integer 
- BorrowerAPR contains 25 NaN values which limits plotting capabilities, should be dropped

**Code:**

```python
df_loan.EmploymentStatus.fillna(value = 'Not available', inplace = True)
```

```python
df_loan.IncomeRange.replace('Not employed', '$0', inplace = True)
```

```python
df_loan.TotalCreditLinespast7years = df_loan.TotalCreditLinespast7years.astype('Int64')
```

```python
df_loan.dropna(subset=['BorrowerAPR'], inplace=True)
```

## Univariate Exploration

Let's start our exploration by looking at the main variables which are the subject of the questions: LoanStatus, BorrowerAPR and LoanOriginalAmount. Are their distribution skewed or symmetric? Are they unimodal or multimodal? How about the categorical data; LoanStatus?

```python
# univariate plot of LoanStatus as a Categorical data, sorted by the highest freqquency 
sns.countplot(df_loan['LoanStatus'], order = df_loan.LoanStatus.value_counts().index ,color = '0.5' )
plt.xticks(rotation = 90);
```

**LoanStatus** Comments:
- Almost all the data can be placed in the first four categories (Current, Completed, Chargedoff, Defaulted).
- Current category contains more loans than any others.

```python
# univariate plot of BorrowerAPR
# Median and Mean comparison 
median_APR= df_loan.BorrowerAPR.median()
mean_APR= df_loan.BorrowerAPR.mean()

# Cut the window in 2 parts
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
 
# Add a graph in each part
sns.boxplot(df_loan["BorrowerAPR"], ax=ax_box)
#sns.distplot(df_loan["BorrowerAPR"], ax=ax_hist)

sns.distplot(df_loan['BorrowerAPR'],bins=30,kde=True).annotate('''Median = {}
Mean = {}'''.format(median_APR, mean_APR), (0.3,4.3))
plt.axvline(x=median_APR, color='red')
plt.axvline(x=mean_APR, color='black');

# Remove x axis name for the boxplot
ax_box.set(xlabel='');
```

**BorrowerRate** Distribution Comments:
- All BorrowerAPR values  larger than 0.45 can be considered as outliers, based on the Box plot
- Mean and Median are very close to each other, Mean is slightly larger than Median
- It is a slightly right skewed distribution with a spike around 0.35

```python
# univariate plot of LoanOriginalAmount 
# Median and Mean comparison 
median_LOA= df_loan.LoanOriginalAmount.median()
mean_LOA= df_loan.LoanOriginalAmount.mean()

# Cut the window in 2 parts
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
 
# Add a graph in each part
sns.boxplot(df_loan["LoanOriginalAmount"], ax=ax_box)

sns.distplot(df_loan['LoanOriginalAmount'],bins=30,kde=True).annotate('''Median = {}
Mean = {}'''.format(median_LOA, mean_LOA), (20000,0.00012))
plt.axvline(x=median_LOA, color='red')
plt.axvline(x=mean_LOA, color='black');

# Remove x axis name for the boxplot
ax_box.set(xlabel='');
```

```python
# univariate plot of LoanOriginalAmount 
# Cut the window in 2 parts
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
 
# Add a graph in each part
mybins=np.logspace(0, np.log(100), 100)
sns.boxplot(df_loan["LoanOriginalAmount"], ax=ax_box)
sns.distplot(df_loan["LoanOriginalAmount"], ax=ax_hist, bins=mybins)
mybins=np.logspace(0, np.log(100), 100)

# Logarithmic x axis scale
plt.xscale('log')

 
# Remove x axis name for the boxplot
ax_box.set(xlabel='');
```

**LoanOriginalAmount** Distribution Comments:
- All values  larger than 25000 can be considered as outliers, based on the Box chart
- Mean and Median are quite different, Mean is larger than Median, indicating positive skewness
- Data shows positively skewed distribution with majority of datapoints falling between 1,000 and 10,000 
- Based on the logarithmic view, data is quite spiky and it is very hard to call it as Unimodal or even Bimodal

## Bivariate Exploration

Above, we looked at the univariate distribution of three features in the **Prosper Loan** dataset: LoanStatus, BorrowerAPR and LoanOriginalAmount. Here, we'll investigate relationships between pairs of these variables, particularly how each of them relate to the three above mentioned features.


```python
# BorrowerAPR variation for each LoanStatus category, arranged by maximum frequency of LoanStatus categories
sns.boxplot(data=df_loan, x='LoanStatus', y='BorrowerAPR', color='0.5', order=df_loan.LoanStatus.value_counts().index)
plt.xticks(rotation=90);
```

**BorrowerAPR for LoanStatus categories** insights:

- Considering the bar chart from univariate exploration of LoanStatus, we know that major LoanStatus categories are established to be Current, Completed, Chargedoff and Defaulted. We use the same order to focus on the characteristics of these major groups.
- Interestingly enough, mean values of BrrowerAPR for Current, Completed LoanStatus categories are significantly lower than Chargedoff and Defaulted categories. 

```python
# BorrowerAPR variation versus BorrowRate
sns.jointplot('BorrowerRate','BorrowerAPR',data=df_loan).annotate(stats.pearsonr);
```

**BorrowerAPR vs. BorrowerRate** insight:

- BorrowAPR is related to Borrower rate. 
- There is a strong positive relationship between the 2 variables. Pearson r of 0.99 is indicative of this fact.

## Multivariate Exploration

So far, we looked at various univariate and bivariate relationships. in the following section, we will look into a few multivariate plots, trying extract more information from our dataset.


```python
# BorrowerAPR versus AvailableBankcardCredit for different CreditGrades


sns.scatterplot(y='BorrowerAPR', x='AvailableBankcardCredit', data=df_loan, 
                hue='CreditGrade');
plt.xscale('log')
plt.xlim(1, 1000000);
```

**BorrowerAPR versus AvailableBankcardCredit for different CreditGrades** :

- There is a very large variation in the BorrowerAPR versus AvailableBankcardCredit. Using the logarithmic scale and categorising based off CreditGrade , creates more insight into the data.
- Although the plot is a little crowded, we can see that similar CreditGrades are clustered near each other. 
- Also, we can see that `HR`, `D` and `E` categories are leaning towards the upper left of the chart, having higher BorrowerAPR while better CreditGrade of `AA` occupies the bottom right corner of the chart which is associated with the lowest possible BorrowerAPR



```python
# creating a pivot table to analyse multiple variables with a heatmap

pivot_table=df_loan.pivot_table(values='BorrowerAPR',index='EmploymentStatus',columns='ProsperRating (numeric)')
pivot_table
```

```python
# Heatmap frop the pivot table
sns.heatmap(pivot_table,cmap='coolwarm'
            ,linecolor='white',linewidths=1, annot=True);
```

**Learning from multivariate heatmap** :

- There is a general trend from left to right of the heatmap, increasing values of `BorrowerAPR` by decreasing `ProsperRating (numeric)`
- While there is not a huge vertical variation of values inside each ProsperRating (numeric) category, `Not employed` group stands out by showing a commonly higher `BorrowerAPR` than other employmentStatus




```python
# BorrowerAPR variation  based on Investors categorized by Term

sns.scatterplot(y='BorrowerAPR', x='Investors', data=df_loan, hue='Term');
```

```python
# LP_CustomerPayments variation  based on LoanOriginalAmount categorized by Term

sns.scatterplot(x='LoanOriginalAmount', y='LP_CustomerPayments', data=df_loan[df_loan['LoanStatus'] =='Current'],
                hue='Term');
```

```python
# BorrowerAPR variation  based on LoanOriginalAmount categorized by Term


sns.scatterplot(x='LoanOriginalAmount', y='BorrowerAPR', data=df_loan, hue='ProsperRating (numeric)');
```

```python
# Scatter plot of Loan Amount on Monthly Payments,for different loan Terms

sns.lmplot(y='MonthlyLoanPayment',x='LoanOriginalAmount',data=df_loan, hue = 'Term', palette='coolwarm'
           ,aspect=2);
plt.ylim(0,2500)
plt.xlim(0,50000)
plt.title('''Effect of Loan Amount on Monthly Payments, 
for different loan Terms''');
```

**Learning from multivariate heatmap** :

- There is a strong positive linear relationship between `LoanOriginalAmount` and `MonthlyLoanPayment` for each different `Term` category
- As expected, higher LoanOriginalAmount increases the MonthlyLoanPayment. 
- The shorter is the term of Loan, the higher becomes the MonthlyLoanPayment
